## Fonction exponentielle et logarithme

Développement parallèle de ces deux fonctions : l'une pour permettre de transformer des multiplications en addition ; l'autre pour rendre compte de croissance spécifiques.

Exponentielle et logarithme sont récirpoques, ce qui signifie que $\exp(\ln(x)) = x$ et $\ln(\exp(x))=x$ (la composition des deux fonctions est l'identité, la fonction x qui à x associe x)

### Exponentielle

La fonction exponentielle est une fonction f notée $\exp$ définie sur R telle que :

$$ f(x) = \exp(x) = e^x$$

Mais il y a quoi dans $\exp$ ? En fait, la valeur exponentielle est le résultat d'une suite $e^x = \lim_{n=+\infty{}}(1+\frac{x}{n})^n$

Cette fonction à certaines propriétés.

Soit a et b deux réels :


- $e^0=1$
- $e^x > 0$
- $\lim_{x=+\infty{}}e^x = +\infty{}$
- $\lim_{x=-\infty{}}e^x = 0$

- $e^{a + b} = e^a.e^b$
- $e^{a-b} = \frac{e^a}{e^b}$
- $(e^a)^b = e^{a.b}$

La fonction exponentielle est dérivable et 

$$ \exp'(x) =  \exp(x) $$

La conséquence quand on a une fonction composée :

$$ \exp'(g(x)) =  g'(x).\exp((g(x)) $$



soit $f(x) = 10.\exp(x^2)$, quelle est sa dérivée ?

C'est la même chose que f(x) = $10.\exp(g(x))$ avec $g(x)=x^2$

Donc $f'(x) = g'(x).f(x) = 2*x*10*\exp(x^2) = 20*x*\exp(x^2)$

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,2,1000)
y = np.exp(x)

fix,ax = plt.subplots(figsize=(20,10))
ax.plot(x,y,".:")
plt.xlabel("X")
plt.ylabel("Y")
```

### Quelques remarques générales

- les fonctions exponentielles sont souvent des solutions d'équations différentielles impliquant une croissance proportionnelle à la valeur d'une fonction à un moment ;
- les fonctions logarithmes permettent de faire des transformations intéressantes passant d'une multiplication à une somme, ce qui historiquement a eu un rôle important quand les ordinateurs n'étaient pas disponibles



### Logarithme

C'est la fonction réciproque (bijective) le logarithme népérien. Elle est définie sur $R^+_*$ et appellée $\ln$

Comme c'est la réciproque de exponentielle, on a $ \ln(e^x) = x$ et $e^{\ln(x)} = x$ ou écrit autrement $\exp ○ \ln = Id$

Soit a et b deux réels :

- $\ln(1) = 0$
- $\lim_{x=+\infty{}}\ln(x) = +\infty{}$
- $\lim_{x=0}\ln(x) = -\infty{}$
- $\ln(a.b) = \ln(a) + \ln(b)$
- $\ln(\frac{a}{b}) = \ln(a) - \ln(b)$
- $\ln(\frac{1}{a}) = - \ln(a)$
- $\ln(a^b) = b.\ln(a)$

- $a^b = e^{\ln{(a)}.b}$

Parce que $e^{\ln{(a)}*b} = (e^{\ln{(a)}})^b = a^b$



Et la dérivée est :

$$ ln'(x) = \frac{1}{x}$$

$$\ln(g(x))' = \frac{g'(x)}{x}$$


Par exemple, $x^{10} = e^{10*\ln(x)}$




### Logarithme décimal


On définit souvent une fonction de logarithme décimal, de la manière suivante :

$$ f(x) = \log_{10}(x) = \frac{\ln(x)}{\ln(10)}$$

Pourquoi ? Parce que cette fonction permet d'avoir la propriété suivante :

Si on prend x, quelle est la valeur du $log_{10}$ de 10*x :

$$  \log_{10}(10 * x) = \frac{\ln(10.x)}{\ln(10)} = \frac{\ln(x)}{\ln(10)} + \frac{\ln(10)}{\ln(10)} = \frac{\ln(x)}{\ln(10)} + 1 = \log_{10}(x) + 1$$

Multiplier par 10 revient à juste ajouter une unité. Ce qui signifie que vous allez avoir une courbe similaire mais décalée vers le haut (+1) si vous étudiez un phénomène qui a une variation proportionnelle mais juste plus amplifiée.

$ln(\frac{a}{a}) = ln(\frac{1}{a}*a) = ln(1) = 0 = ln(a) + ln(\frac{1}{a})$

$$f(x) = \exp(10*x+5)$$

$$\ln(f(x)) = 10*x+5$$

## Régression linéaire

La philosophie est très simple : vous vous intéressez à une grandeur (souvent notée Y) et vous aimeriez mieux comprendre comment sa variation dépend d'un ensemble de facteurs. Dans le cas où vous avez de bonne raison de penser qu'une autre grandeur (X) est responsable de sa variation (quand X varie, alors Y varie), vous voulez écrire cette relation.

La forme la plus simple qui relie deux grandeurs est la relation linéaire : si on a une variation $\Delta X$ alors Y varie aussi, d'une quantité $\Delta Y$, et la relation entre ces deux valeurs est $\Delta Y = a.\Delta X$

Dit autrement, la manière la plus simple de relier Y à X est de faire l'hypothèse que les deux sont reliés par une équation de droite.

Or, ce n'est jamais tout à fait vrai. Vos données ne sont jamais reliées par une relation exacte linéaire. C'est une hypothèse. Et même s'il y a bien une "vraie" relation linéaire entre vos données, le hasard et les erreurs de mesure font que les points ne sont pas exactement là où vous les attendez.

L'objectif est alors de trouver la meilleure droite qui permet de s'adapter à vos mesures : **c'est ça une régression linéaire**.

Comment ? L'idée est de prendre toutes les droites possibles et de garder celle qui est la plus proche des données réelles.

### Avec des données

```python
X = [3,20,2,78,56,9,38,5,100]
Y = [8,34,1,56,13,56,10,4,78]

fix,ax = plt.subplots(figsize=(10,5))
ax.plot(X,Y,"o")
plt.xlabel("X")
plt.ylabel("Y")
```

Définisson une droite au hasard

```python
a = 1
b = 3
Y_d = [a*x + b for x in X] + [b]

fix,ax = plt.subplots(figsize=(10,5))
ax.plot(X,Y,"o")
ax.plot(X + [0],Y_d)

plt.xlabel("X")
plt.ylabel("Y")
```

On peut montrer par le calcul que si on cherche la droite qui rend l'erreur quadratique la plus petite (la somme des carrés des distances entre la droite et le point), on a une expression 

$$  b = \frac{ \sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y}) }{ \sum_{i=1}^n (x_i - \bar{x})^2 } $$
 
 et 
 
 $$  a = \bar{y} - b.\bar{x}
$$

Dans les faits, les logiciels calculent directement ces valeurs

```python
# Calculer la droite de régression
from scipy import stats
a, b, r_value, p_value, std_err = stats.linregress(X,Y)


# Visualiser la droite

Y_d = [a*x + b for x in X] + [b]

fix,ax = plt.subplots(figsize=(10,5))
ax.plot(X,Y,"o")
ax.plot(X + [0],Y_d)

plt.xlabel("X")
plt.ylabel("Y")
```
